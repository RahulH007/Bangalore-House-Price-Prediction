## Handeling Total_Size column

```python
data['total_sqft'].unique()
```

```python
def convertrange(n):
    temp=n.split('-')
    if temp==2:
        return (float(temp[0])+float(temp[1]))/2
    try:
        return float(n)
    except:
        return None
```

```python
data['total_sqft']=data['total_sqft'].apply(convertrange) # Changing Range Scale to Average
```

# Outlier Detection & Removal

```python
data.describe()
```

```python
data=data[((data['total_sqft']/data['bhk'])>=300)]  #To Remove Min 1sqft total_size
data.describe()
```

```python
data.price_per_sqft.describe()
```

```python
def remove_outlier_sqft(df):
    df_output=pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m=np.mean(subdf.price_per_sqft)
        st=np.std(subdf.price_per_sqft)

        gen_df=subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_output=pd.concat([df_output,gen_df],ignore_index=True)
    return df_output
data=remove_outlier_sqft(data)
data.describe()
```

```python
def bhk_outlier_removal(df):
    ex_ind=np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats={}
        for bhk ,bhk_df in location_df.groupby('bhk'):
            bhk_stats={
                'mean': np.mean(bhk_df.price_per_sqft),
                'std':np.std(bhk_df.price_per_sqft),
                'count':bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                ex_ind=np.append(ex_ind,bhk_df[bhk.price_per_sqft<(stats['mean'])].index.values) 
    return df.drop(ex_ind,axis='index')
```

```python
data=bhk_outlier_removal(data)
```
